## 4.知识小结

### 模块的导入

import语句导入模块
```python
import math
print(math.sqrt(a))
```
from import语句导入指定函数
```python
from math import sqrt 
print(sqrt(a))
```
用\*导入全部函数。
```python
from math import * 
print(sqrt(a))
```
